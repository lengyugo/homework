## 2.Supervised Direction to get optimal k* and *b

````python
try_times = 2000
min_loss = float('inf')
best_k = random.random() * 200 -150
best_b = random.random() * 200 -150
direction = [(+1,-1),(+1,+1),(-1,-1),(-1,+1)]

next_direction = random.choice(direction)
update_times = 0
scaler = 0.1

for i in range(try_times):
    k_direction,b_direction = next_direction
    current_k,current_b = best_k + k_direction * scaler,best_b + b_direction * scaler 
    
    price_k_and_b = [price(r,current_k,current_b)for r in X]
    current_loss = loss(y,price_k_and_b)
    
    if current_loss < min_loss:
        min_loss = current_loss
        
        best_k,best_b = current_k,current_b
        
        next_direction = next_direction
        update_times += 1
        
        if update_times % 10 == 0:
            print('when times is :{},get best_k: {} best_b: {},and the loss is {}'.
                  format(i,best_k,best_b,current_loss))
    else:
        next_direction = random.choice(direction)
````

## 3.Gradient Descent to get optimal k* and *b

````python
def partial_k(x,y,y_hat):
    n = len(y)
    
    gradient = 0
    for x_i,y_i,y_hat_i in zip(list(x),list(y),list(y_hat)):
        gradient += (y_i - y_hat_i)* x_i
        
    return -2/n * gradient

def partial_b(x,y,y_hat):
    n = len(y)
    
    gradient = 0
    for x_i,y_i,y_hat_i in zip(list(x),list(y),list(y_hat)):
        gradient += (y_i - y_hat_i)
        
    return -2/n * gradient
````

````python
try_times = 2000
min_loss = float('inf')

current_k = random.random() * 200 - 100
current_b = random.random() * 200 - 100

update_times = 0
learn_rate = 1e-04

for i in range(try_times):

    price_k_and_b = [price(r,current_k,current_b)for r in X]
    current_loss = loss(y,price_k_and_b)
    
    if current_loss < min_loss:
        
        min_loss = current_loss
    
        update_times += 1
        
        if update_times % 50 == 0:
            print('when times is :{},get best_k: {} best_b: {},and the loss is {}'.
                  format(i,current_k,current_b,current_loss))
    g_greadint = partial_k(X,y,price_k_and_b)
    k_greadint = partial_b(X,y,price_k_and_b)
    
    current_k = current_k + (-1 * g_greadint * learn_rate)
    current_b = current_b + (-1 * k_greadint * learn_rate)
````

###### 1. Why do we need machine learning methods instead of creating a complicated formula?


Ans:因为有很多复杂的问题不能创造出一个公式来解决问题，或者很难找到一个符合的公式。

###### 2.  Wha't's the disadvantages of `the 1st Random Choosen` methods in our course? 

Ans:1.加大了计算数据量，浪费资源。2.完全随机的尝试增加了很多不必要的工作。

###### 3. Is the `2nd method supervised direction` better than 1st one?  What's the disadvantages of `the 2nd supversied directin` method? 

Ans:第二种方法要比第一种好很多，在同样的尝试次数下，第二种方法可以达到更好的效果。
我们希望损失函数越小越好，第二种方式不能够达到我们的预期。

###### 4. Why do we use `Derivative / Gredient` to fit a target function? 

Ans:梯度的几何意义就是函数变化最快的方向，我们沿着梯度的反方向，梯度减少的最快，这样可以找到函数的最小值。而对于我们这个问题来说，找出梯度的方向，然后沿着这个方向去变化，是最快的找到符合我们预期的方法。

###### 5. In the words 'Gredient Descent', what's the `Gredient` and what's the `Descent`?

Ans:梯度二元函数中就是对参数求偏导数，在多元函数中就是把求得的各个参数的偏导数以向量的形式表示出来就是梯度。
而下降的意思是，对于一个多维的函数来说，我们不可能一下子找到损失函数的最小值，我们可以先一个维度上找到损失函数的最小值，然后一步一步的的找到整个函数的损失函数的最小值。

###### 6. What's the advantages of `the 3rd gradient descent method` compared to the previous methods?

Ans:梯度下降保证了我们的尝试一定是在正确的方向上收敛的，可以得出比较好的效果。

###### 7. Using the simple words to describe: What's the machine leanring.

Ans:机器学习就是一种，通过不断地输入数据量，模型参数会逐渐朝着我们想要的方向变化的一种方式。
