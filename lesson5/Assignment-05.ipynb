```python
from collections import defaultdict
from functools import lru_cache
```

```python
##加动态规划
seen = defaultdict(int)

#@lru_cache(maxsize=2**10)

def get_relate_word(initial_word,model):
    
    unseen = initial_word
    
    max_size = 500
    while unseen and len(seen) < max_size:
        
        if len(seen) % 50:
            print('seen length {}'.format(len(seen)))
        
        node = unseen.pop(0)
        
        new_expanding = [w for w,f in model.wv.most_similar(node,topn=20)]
        
        unseen += new_expanding
        
        seen[node] += 1 ##改变加权方式
        
    return seen
```

```python
get_relate_word(['说','告诉'],model)
```

```python
content = pd.read_csv(csv_path,encoding='gb18030')
```

```python
import jieba
```

```python
def cuting(string):
    return ' '.join(jieba.cut(string))
```

```python
cuting('这是一个测试')
```

```python
import re
```

```python
def token(string):
  
    return ''.join(re.findall(r'[\d|\w]+',string))
```

```python
token('此外，自本周（6月12日）起，除小米手机6等15款机型外，其余机型已暂停更新发布（含开发版/体验版内测，稳定版暂不受影响），以确保工程师可以集中全部精力进行系统优化工作。有人猜测这也是将精力主要用到MIUI 9的研发之中。\r\nMIUI 8去年5月发布，距今已有一年有余，也是时候更新换代了。\r\n当然，关于MIUI 9的确切信息，我们还是等待官方消息。\r\n')
```

```python
new_content = []
for n,line in enumerate(news_content):
    if n < 35:
        new_content.append(token(line))
  
```

```python
new_content = [cuting(n) for n in new_content]
```

```python
new_content[1]
```

```python
with open('cut_new_content.txt','w',encoding='utf-8') as f:
    for n in new_content:
        f.write(n + '\n')
```
